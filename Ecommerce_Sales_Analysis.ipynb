```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

Task 1

```python
print("--- TASK 1 ---")

# 1. Total sales amount for each region (sorted descending)
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()
region_sales_sorted = region_sales.sort_values(by='sales_amount', ascending=False)
print("\nRegional Sales (Sorted):\n", region_sales_sorted)

# 2. Count the number of transactions for each product category
category_counts = df.groupby('category')['transaction_id'].count().reset_index()
print("\nTransactions per Category:\n", category_counts)

# 3. Average sales amount per salesperson
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()
print("\nAverage Sales per Salesperson:\n", salesperson_avg)
```

Task 2

```python
print("\n--- TASK 2 ---")

# 1. Group by region AND category
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()
print("\nSales by Region and Category:\n", region_category_sales.head()) # Showing top 5 rows for brevity

# 2. Salesperson metrics (sum, mean, count) sorted by total sales
salesperson_metrics = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()
salesperson_metrics_sorted = salesperson_metrics.sort_values(by='sum', ascending=False)
print("\nSalesperson Metrics (Sorted by Total Sales):\n", salesperson_metrics_sorted)
print(f"Top Performer: {salesperson_metrics_sorted.iloc[0]['salesperson']}")

# 3. Maximum total revenue by category using .idxmax()
# Note: We don't use reset_index() here because idxmax() works on the Series index
category_totals = df.groupby('category')['sales_amount'].sum()
top_category = category_totals.idxmax()
print(f"\nCategory with Maximum Revenue: {top_category}")
```

Task 3

```python
print("\n--- TASK 3 ---")

# 1. Define custom aggregation function
def sales_range(x):
    return x.max() - x.min()

# 2. Apply custom function with standard aggregations
region_stats = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
print("\nDetailed Regional Stats (with Custom Range):\n", region_stats)

# 3. Final summary report using dictionary syntax
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print("\nFinal Summary Report:\n", final_report)
```
